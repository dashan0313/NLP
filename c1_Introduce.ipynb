## 词出现频率

```python
fdist1 = FreqDist(text1) #类似字典
```

```python
fdist1.most_common(50)
```

```python
fdist1.plot(50,cumulative=True)
```

```python
#仅出现一次的词
fdist1.hapaxes()
```

## 细粒度选词

```python
V = set(text1)
long_words = [w for w in V if len(w) > 15]
sorted(long_words)
```

```python
#复杂点的任务：找出又长出现次数又多的词
fdist5 = FreqDist(text5)
sorted(w for w in set(text5) if len(w)>7 and fdist5[w]>7)
```
